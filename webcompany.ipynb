```
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```
#List of compnay from web
company_name = []
company_ticker = []
```

```
URL = 'https://www.advfn.com/nyse/newyorkstockexchange.asp?companies=A'
page = requests.get(URL)
soup = BeautifulSoup(page.text, 'html.parser')
odd_rows = soup.find_all('tr', attrs={'class':'ts0'})
evens_rows = soup.find_all('tr', attrs={'class':'ts1'})
```

```
for i in odd_rows:
  row = i.find_all('td')
  company_name.append(row[0].text.strip())#List of Company name
  company_ticker.append(row[1].text.strip())#List of Ticker

for i in evens_rows:
  row = i.find_all('td')
  company_name.append(row[0].text.strip())#List of Company name
  company_ticker.append(row[1].text.strip())#List of Ticker 
```

```
def scrape_stock_symbols(Letter):
  Letter = Letter.upper()
  URL = 'https://www.advfn.com/nyse/newyorkstockexchange.asp?companies='+Letter
  page = requests.get(URL)
  soup = BeautifulSoup(page.text, 'html.parser')
  odd_rows = soup.find_all('tr', attrs={'class':'ts0'})
  evens_rows = soup.find_all('tr', attrs={'class':'ts1'})

  for i in odd_rows:
    row = i.find_all('td')
    company_name.append(row[0].text.strip())#List of Company name
    company_ticker.append(row[1].text.strip())#List of Ticker

  for i in evens_rows:
    row = i.find_all('td')
    company_name.append(row[0].text.strip())#List of Company name
    company_ticker.append(row[1].text.strip())#List of Ticker 

  return (company_name, company_ticker)
```

```
(temp_name, temp_ticker) = scrape_stock_symbols('m')
temp_name
```

```
import string
string.ascii_uppercase
```

```
for char in string.ascii_uppercase:
  (temp_name, temp_ticker) = scrape_stock_symbols(char)
```

```
data = pd.DataFrame(columns=['company_name', 'company_ticker'])
data['company_name'] = temp_name
data['company_ticker'] = temp_ticker
data
```
